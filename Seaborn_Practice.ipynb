### 폰트 설정
그래프에 한글이 있을땐, os별 기본 한글 폰트 설정 필요
1. Windows
set_theme(style='white', font='Malgun Gothic')

2. Mac
set_theme(style='white', font='AppleGothic')

폰트 크기는 font_scale. 이 파라미터는 원래 기본값이 1이라서 0.5, 0.8 같이 1보다 작은 값을 주면 폰트 크기가 작아지고, 1.5, 2 같이 1보다 큰 값을 주면 폰트 크기가 커짐

```python
sns.set_theme()
sns.barplot(data = df, x='month', y='total', errorbar=None)
plt.show()
```

#### 씨본 스타일 - 팔레트 파라미터

이 파라미터의 기본값은 deep. 이것 말고도 pastel, muted, bright, dark, colorblind 등이 있음.

<명령어>

sns.color_palette: 색상 리스트를 불러내줘

sns.palplot: 어떤 색인지 직접 보여줘

```python
sns.palplot(sns.color_palette("tab10"))
sns.palplot(sns.color_palette("hls"))
```

```python
sns.set_palette(sns.color_palette()) ## 가장 디폴트 색
sns.barplot(data = df, x='month', y='total', errorbar=None)
plt.show()
```

```python
sns.set_theme(palette='muted') ## theme 함수 이용한 사용법
sns.barplot(data = df, x='month', y='total', errorbar=None)
plt.show()
```

```python
## 직접 색 불러와 지정
colors = ["#33FF33"]
sns.set_palette(sns.color_palette(colors))
sns.barplot(data = df, x='month', y='total', errorbar=None)
plt.show()
```

```python
## theme 함수에 여러파라미터 지정
sns.set_theme(style='white', font='Malgun Gothic', palette='pastel', rc={'figure.figsize': (12, 6)})
sns.barplot(data = df, x='month', y='total', errorbar=None)
plt.show()
```
